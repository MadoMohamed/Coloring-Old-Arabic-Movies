```python
import cv2
video = cv2.VideoCapture('My/Dalila Movie.mp4')
success,image = video.read()
frame_no = 0
img_no = 0
print(success)
while success:
    if (frame_no%175==0):
        print ('Read a new frame: ', success)
        image = cv2.resize(image,(256,256))
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        if(img_no>15):
            cv2.imwrite("My/inputs/frame%d.jpg" % (img_no-16), gray)     # save input frame as JPEG file
            cv2.imwrite("My/outputs/label%d.jpg" % (img_no-16), image)     # save output frame as JPEG file
        img_no+=1
    success,image = video.read()
    frame_no += 1
print("done ...")
```

```python
import pickle
import os
import numpy as np

pickle_input = open("My/inputs.p","wb")
no_images = input('enter no of images: ')
name_input_file = input('enter name of folder that holds images: ')
frame_name = input('enter frame name: ')
name_lable_file = input('enter name of folder that holds labels: ')
label_name = input('enter label name: ')

list_images = []
for i in range(int(no_images)):
    image = cv2.imread("My"+'/'+name_input_file + '/' + frame_name + str(i) + '.jpg')
    list_images.append(image)

pickle.dump(list_images, pickle_input)
pickle_input.close()

pickle_out = open("My/outputs.p","wb")

list_labels = []
for i in range(int(no_images)):
    label = cv2.imread("My"+'/'+name_lable_file + '/' + label_name + str(i) + '.jpg')
    list_labels.append(label)

pickle.dump(list_labels, pickle_out)    
pickle_out.close()

print('finish')
```
