```python
import cv2
vidcap = cv2.VideoCapture('/home/mohamed/Downloads/movie1.mp4')
success,image = vidcap.read()
count = 0
count2 = 0
print(success)
while success:
    if (count%150==0):
        print ('Read a new frame: ', success)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        gray = cv2.resize(gray,(256,256))
        cv2.imwrite("/home/mohamed/Downloads/inputs/frame%d.jpg" % count2, gray)     # save input frame as JPEG file
        rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        rgb = cv2.resize(rgb,(256,256))
        cv2.imwrite("/home/mohamed/Downloads/outputs/label%d.jpg" % count2, rgb)     # save output frame as JPEG file
        count2+=1
    success,image = vidcap.read()
    count += 1
print("done ...")
```

```python
import cv2
import pickle
import os
import numpy as np

pickle_out1 = open("inputs.p","wb")
no_images = input('enter no of images: ')
name_input_file = input('enter name of folder that holds images: ')
frame_name = input('enter frame name: ')
name_lable_file = input('enter name of folder that holds labels: ')
label_name = input('enter label name: ')

list_images = []
count = 0
for i in range(int(no_images)):
    image = cv2.imread("/Downloads"+'/'+name_input_file + '/' + frame_name + str(i) + '.jpg',0)
    list_images.append(image)

pickle.dump(list_images, pickle_out1)
pickle_out1.close()

pickle_out2 = open("outputs.p","wb")

list_labels = []
for i in range(int(no_images)):
    label = cv2.imread("/Downloads"+'/'+name_lable_file + '/' + label_name + str(i) + '.jpg',1)
    list_labels.append(label)

pickle.dump(list_labels, pickle_out2)    
pickle_out2.close()

print('finish')
    
    
```
